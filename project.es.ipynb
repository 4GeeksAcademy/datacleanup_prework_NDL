No es posible "eliminar" todos los NAs de todas las columnas. Por ejemplo, con dropna() se eliminan filas completas y se pierde mucha información.
Si se busca imputar con media, mediana o moda, no resulta apropiado para números de telefono, nombres de calles, etc.
Al ver las variables numéricas que tendría sentido imputar, se encuentran sesgadas, por lo tanto, usaré la mediana.
Todos los demás nulos los dejaré.

```python
for columna in columnas_imputar:
    mediana = ds[columna].median()
    ds[columna].fillna(mediana, inplace=True)
    
ds.isnull().sum().sort_values(ascending=False).head(9)
```

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, Marker, Circle, basemaps, LayersControl

# Crear un diccionario para asignar colores a cada población
color_dict = {'Fuenlabrada': 'red', 'Leganés': 'blue', 'Getafe': 'green', 'Alcorcón': 'purple'}

# Crear el mapa centrado en Madrid
map = Map(center=(40.4168, -3.7038), zoom=11, basemap=basemaps.OpenStreetMap.Mapnik)

# Agregar marcadores para cada población con su color correspondiente
for index, row in df_cinturon_sur.iterrows():
    marker = Circle(
        location=(row['latitude'], row['longitude']),
        radius=50,
        color=color_dict[row['level5']],
        fill=True,
        fill_color=color_dict[row['level5']]
    )
    map.add_layer(marker)

# Agregar un control de capas para mostrar/ocultar marcadores
map.add_control(LayersControl())

# Mostrar el mapa
map
```
